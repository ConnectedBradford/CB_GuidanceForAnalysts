```python
from google.cloud import bigquery
import pandas as pd
import pandas_gbq

# Initialize BigQuery client
client = bigquery.Client(project="yhcr-prd-phm-bia-core")

# Databases
source_db = "yhcr-prd-phm-bia-core.CB_STAGING_DATABASE_WAREHOUSE_FDM_Format"
target_db = "yhcr-prd-phm-bia-core.CB_MYSPACE_PAA"

# Specific tables to check
tables_to_check = [
    "tbl_maternity_pathway_antenatal",
   "tbl_autism_amalgamated_ptl_oct2022",
     "tbl_spell"
]

def get_row_count(dataset_id, table_name):
    """Returns the number of rows in the specified table."""
    query = f"SELECT COUNT(*) as row_count FROM `{dataset_id}.{table_name}`"
    query_job = client.query(query)
    results = query_job.result()
    for row in results:
        return row.row_count

# Loop through each table and compare row counts
for table in tables_to_check:
    source_count = get_row_count(source_db, table)
    target_count = get_row_count(target_db, table)
    print(f"Table: {table}, Source Rows: {source_count}, Target Rows: {target_count}")
    if source_count == target_count:
        print("Row counts match.")
    else:
        print("Row counts do not match.")
```

```python
# Check without using specific table names.
from google.cloud import bigquery

# Initialize BigQuery client
client = bigquery.Client(project="yhcr-prd-phm-bia-core")

# Databases
source_db = "yhcr-prd-phm-bia-core.CB_STAGING_DATABASE_WAREHOUSE_FDM_Format"
target_db = "yhcr-prd-phm-bia-core.CB_FDM_BradfordRoyalInfirmary"

def get_table_names(dataset_id):
    """Returns a list of table names in the given dataset."""
    tables = client.list_tables(dataset_id)
    return [table.table_id for table in tables]

def get_row_count(dataset_id, table_name):
    """Returns the number of rows in the specified table."""
    query = f"SELECT COUNT(*) as row_count FROM `{dataset_id}.{table_name}`"
    query_job = client.query(query)
    results = query_job.result()
    for row in results:
        return row.row_count

# Get table names from source and target datasets
source_tables = get_table_names(source_db)
target_tables = get_table_names(target_db)

# Loop through each table and compare row counts
for table in source_tables:
    if table in target_tables:
        source_count = get_row_count(source_db, table)
        target_count = get_row_count(target_db, table)
        print(f"Table: {table}, Source Rows: {source_count}, Target Rows: {target_count}")
        if source_count == target_count:
            print("Row counts match.")
        else:
            print("Row counts do not match.")
    else:
        print(f"Table {table} does not exist in target database.")
```
